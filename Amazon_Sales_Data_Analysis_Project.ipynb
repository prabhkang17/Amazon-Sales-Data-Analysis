```python
import pandas as pd

data = pd.read_csv('amazon_sales_data.csv')

data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['discounted_price'] = data['discounted_price'].str.replace('₹','')
data['discounted_price'] = data['discounted_price'].str.replace(',','')
data['discounted_price'] = data['discounted_price'].astype(float)
```

```python
data['actual_price'] = data['actual_price'].str.replace('₹','')
data['actual_price'] = data['actual_price'].str.replace(',','')
data['actual_price'] = data['actual_price'].astype(float)
```

```python
data['discount_percentage'] = data['discount_percentage'].str.replace('%','')
data['discount_percentage'] = data['discount_percentage'].astype(float)
```

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
```

```python
average_rating = data['rating'].mean()

print("Average Rating:", average_rating)
```

```python
max_discount = data['discount_percentage'].max()

print("Maximum Discount:", max_discount)
```

```python
avg_price = data['actual_price'].mean()

print("Average Actual Price:", avg_price)
```

```python
top_rated = data[['product_name','rating']].sort_values(by='rating', ascending=False)

top_rated.head()
```

```python
category_count = data['category'].value_counts()

print(category_count)
```

```python
top_categories = data['category'].value_counts().head(10)

print(top_categories)
```

```python
top_categories.plot(kind='bar')

plt.title("Top Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.show()
```

```python
data['rating'].plot(kind='hist')

plt.title("Distribution of Ratings")
plt.xlabel("Ratings")

plt.show()
```

```python
top5 = data['category'].value_counts().head(5)

top5.plot(kind='pie', autopct='%1.1f%%')

plt.ylabel('')

plt.title("Top 5 Product Categories")

plt.show()
```

```python
plt.scatter(data['actual_price'], data['rating'])

plt.title("Price vs Rating")
plt.xlabel("Actual Price")
plt.ylabel("Rating")

plt.show()
```

```python
# Finding top 10 most expensive products

expensive = data[['product_name','actual_price']].sort_values(by='actual_price', ascending=False)

expensive.head(10)
```

```python
# Finding products with highest discounts

highest_discount = data[['product_name','discount_percentage']].sort_values(by='discount_percentage', ascending=False)

highest_discount.head(10)
```

```python
# Calculating average rating for each category

category_rating = data.groupby('category')['rating'].mean()

category_rating.head()
```
